```python
from llama_cpp import Llama

llm = Llama.from_pretrained(
	repo_id="TheBloke/TinyLlama-1.1B-Chat-v1.0-GGUF",
	filename="tinyllama-1.1b-chat-v1.0.Q2_K.gguf",
)
```

```python
from llama_cpp import LlamaGrammar
def get_expression_that_solves(math_problem: str) :
    system_prompt = """
  You are a math instructor. I will ask you a math question.
  Respond with the mathematical expression than can be used to solve the problem.
"""
    messages = [
        {"role":"system","content":system_prompt}
        ,{"role":"user","content": math_problem}
    ]
    # load the grammar
    
    grammar_str = """
root  ::= (expr "=" ws term "\n")+
expr  ::= term ([-+*/] term)*
term  ::= ident | num | "(" ws expr ")" ws
ident ::= [a-z] [a-z0-9_]* ws
num   ::= [0-9]+ ws
ws    ::= [ \t\n]*
    """
    grammar = LlamaGrammar(_grammar=grammar_str)
    response = llm.create_chat_completion(
        grammar=grammar,
        messages=messages)
    
    return response['choices']
```

```python
result = get_expression_that_solves("""
Bill has 3 apples and 2 oranges.
Mae has 2 apples and 4 oranges.
How many total apples do Bill and Mae have?
                                    """)
```
